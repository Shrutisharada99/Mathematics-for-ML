# Finding minimum value of a function
## Brute Force Method

```python
iter = 1000
tol = 0.01
delta = 0.001
x0 = 10
```

```python
def eqn(x):
    return (x-3)**2
```

```python
import numpy as np
import matplotlib.pyplot as plt

xn = np.linspace(-14,20)
yn = eqn(xn)
plt.figure(figsize=(15,4))
plt.plot(xn, yn)
plt.show()
```

```python
x0 = 10
iter =  1000
delta = 0.01
for i in range(iter):
    y = eqn(x0)
    y1 = eqn(x0+ delta)
    if y <= tol:
        print(i, 'Root found at:', 'x0:', x0, 'y:', y)
        break
    else:
        if y1 > y:
            x0 = x0 - delta
        else:
            x0 = x0 + delta
```

## Gradient Descent Approach to find minimum

```python
LR = 0.01  # Learning Rate
iter = 1000
tol = 0.001
x0 = 10
```

```python
def eqn(x):
    return (x-3)**2
```

```python
def deqn(x):
    return 2*(x-3)
```

```python
x0 = 10
iter = 1000

for i in range(iter):
    dy_dx = deqn(x0)
    x0 = x0 - (LR*dy_dx)
    y = eqn(x0)
    if (y < tol):
        print(i, "Minimum found at:", x0, 'y:', eqn(x0))
        break
    if i%100 == 0:
        print(i, 'x0:', x0, 'y:', y)
        plt.figure(figsize=(15,4))
        plt.plot(xn,yn)
        plt.scatter(x0, y, color='r')
        plt.draw()
```

### Finding minimum using Gradient Descent Approach (eq2)

```python
LR = 0.01  # Learning Rate
iter = 1000
tol = 0.01
x0 = 10
```

```python
def eqn(x):
    return (x-3)**2
```

```python
def deqn(x):
    return 2*(x-3)
```

```python
xn = np.linspace(-12, 20, 50)
yn = eqn(xn)
```

```python
x0 = 10
iter = 1000

for i in range(iter):
    dy_dx = deqn(x0)
    x0 = x0 - (LR*dy_dx)
    y = eqn(x0)
    if (dy_dx < tol):
        print(i, "Minimum found at:", x0, 'y:', eqn(x0))
        break
    if i%100 == 0:
        print(i, 'x0:', x0, 'y:', y)
        plt.figure(figsize=(15,4))
        plt.plot(xn,yn)
        plt.scatter(x0, y, color='r')
        plt.show()
```
